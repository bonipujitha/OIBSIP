```python
import plotly.express as px
# Sample data in the correct format
data_df = pd.DataFrame({
    "Region": ["Region1", "Region1", "Region2", "Region2"],
    "Subregion": ["Subregion1", "Subregion2", "Subregion3", "Subregion4"],
    "Estimated Unemployment Rate(%)": [5.0, 7.2, 4.8, 3.1]
})

# Create the sunburst chart
figure = px.sunburst(data_df, path=["Region", "Subregion"], 
                     values="Estimated Unemployment Rate(%)",
                     width=500, height=500, color_continuous_scale="RdYlGn",
                     title="Unemployment Rate in India")

# Show the chart
figure.show()
```

```python

# Sample data
data_df = {
    "Region": ["India", "India", "India", "Region1", "Region1"],
    "Subregion": ["State1", "State2", "State3", "City1", "City2"],
    "Estimated Unemployment Rate(%)": [5.0, 7.2, None, 3.1, 6.5]
}

unemployment = pd.DataFrame(data_df)

# Create the heatmap for missing values
plt.figure(figsize=(4, 4))  # Adjust the figure size
sns.heatmap(unemployment.isnull(), cmap="viridis", cbar=False, annot=True, fmt=".0f")
plt.title("Missing Values Heatmap")
plt.show()
```

```python
# Set the plot style and figure size
plt.style.use('seaborn-whitegrid')
plt.figure(figsize=(8, 4))

# Calculate the correlation matrix and plot the heatmap
sns.heatmap(data_df.corr(), annot=True, cmap="coolwarm", fmt=".2f", center=0)

# Set the title and display the plot
plt.title("Correlation Heatmap")
plt.show()
```

```python

# Rename the column "Estimated Labour Participation Rate(%)" to "Estimated Labour Participation Rate"
data_df.rename(columns={"Estimated Labour Participation Rate(%)": "Estimated Labour Participation Rate"}, inplace=True)

# Check the data type of "Estimated Employed"
print(data_df["Estimated Unemployment Rate(%)"].dtype)

# If "Estimated Employed" is not numeric, convert it to numeric
if not pd.api.types.is_numeric_dtype(data_df["Estimated Unemployment Rate(%)"]):
    data_df["Estimated Unemployment Rate(%)"] = pd.to_numeric(data_df["Estimated Unemployment Rate(%)"], errors="coerce")

# Set the plot style
plt.style.use('seaborn-whitegrid')

# Create the histogram
sns.histplot(x="Estimated Unemployment Rate(%)", data=data_df)

# Set the plot title
plt.title("Indian Unemployment")

# Display the plot
plt.show()
```
